```python
# -*- coding: utf-8 -*-

# La librairie numpy contient la majorit√© des √©l√©ments math√©matiques utilis√©s dans ce travail.
import numpy as np
# La librairie matplotlib permet de g√©n√©rer des graphiques rapidement.
import matplotlib.pyplot as plt
# Le module seaborn permet de faire des meilleurs pr√©sentations graphiques.
import seaborn as sns
# Le module cv2 de la librairie OpenCV-Python est con√ßue pour r√©soudre les probl√®mes de vision par ordinateur.
import cv2 as cv
# Le module OS fournit des fonctions pour manipuler un r√©pertoire (dossier).
import os
```

```python
def load_images_from_folder(folder):
    """Fonction servant √† charger des images √† partir d'un dossier.

    Args:
        folder (string): Le chemin d'acc√®s au dossier qui contient les images.

    Returns:
        dictionary: Dictionnaire encodant l'image avec son nom.
    """
    images_dict = {}
    for filename in os.listdir(folder):
        img = plt.imread(os.path.join(folder, filename))
        if img is not None:
            images_dict.update({filename:img})
    return images_dict

def filenames(measures_dict):
    """Fonction permettant d'extraire les noms de fichiers.

    Args:
        measures_dict (dictionary): Dictionnaire d'images trait√©s pr√©alablement
            par la fonction load_images_from_folder.

    Returns:
        tuple: Nom du fichier complet et nom raccourci.
    """
    complete_filenames = list(measures_dict)
    short_filenames = []
    for name in complete_filenames:
        short_filenames.append(name.replace(".jpg", "").replace("_"," "))
    return complete_filenames, short_filenames
```

```python
measures_HeNe = load_images_from_folder("Images\Measures_HeNe")
complete_names_HeNe, short_names_HeNe = filenames(measures_HeNe)
fig, ax = plt.subplots(figsize=(13,12))
ax.imshow(measures_HeNe[complete_names_HeNe[0]])
ax.set_title("{} Image".format(short_names_HeNe[0]), fontsize=20)
ax.set_xlabel("Horizontal pixel scaling", fontsize=16)
ax.set_ylabel("Vertical pixel scaling", fontsize=16)
plt.show()
```
